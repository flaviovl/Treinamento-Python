Q2.O que são namespaces Python?

Resp: Um namespace em python se refere ao nome que é atribuído a cada objeto em python. Os objetos são variáveis ​​e funções. À medida que cada objeto é criado, seu nome junto com o espaço (o endereço da função externa na qual o objeto está), é criado. Os namespaces são mantidos em python como um dicionário onde a chave é o namespace e o valor é o endereço do objeto. Existem 4 tipos de namespace em python-

Namespace interno – Esses namespaces contêm todos os objetos internos em python e estão disponíveis sempre que python estiver em execução.
Espaço de nomes global – Estes são espaços de nomes para todos os objetos criados no nível do programa principal.
Incluindo namespaces – Esses namespaces estão no nível mais alto ou na função externa.
Espaços de nomes locais – Esses espaços de nomes estão na função local ou interna.

Q10. Qual é a diferença entre cópia profunda e superficial?
    Resp:  A cópia superficialé usada quando um novo tipo de instância é criado e mantém os valores que são copiados na nova instância. 
    A cópia rasa é usada para copiar os ponteiros de referência da mesma forma que copia os valores. Essas referências apontam para os objetos 
    originais e as alterações feitas em qualquer membro da classe também afetarão a cópia original da mesma. A cópia rasa permite uma execução 
    mais rápida do programa e depende do tamanho dos dados usados.

    A cópia profunda é usada para armazenar os valores que já foram copiados. A cópia profunda não copia os ponteiros de referência para os 
    objetos. Ele faz a referência a um objeto e o novo objeto que é apontado por algum outro objeto é armazenado. As alterações feitas na cópia 
    original não afetarão nenhuma outra cópia que use o objeto. A cópia profunda torna a execução do programa mais lenta devido a certas cópias 
    para cada objeto que é chamado.

Q11. Explique a herança em Python com um exemplo.
    Resposta:  A herança permite que uma classe ganhe todos os membros (digamos, atributos e métodos) de outra classe. A herança fornece reutilização de código, facilita a criação e a manutenção de um aplicativo. A classe da qual estamos herdando é chamada de superclasse e a classe que é herdada é chamada de classe derivada/filho.

    Eles são diferentes tipos de herança suportados pelo Python:

    Herança Única – onde uma classe derivada adquire os membros de uma única superclasse.
    Herança multinível – uma classe derivada d1 é herdada da classe base base1 e d2 é herdada da base2.
    Herança hierárquica – de uma classe base você pode herdar qualquer número de classes filhas
    Herança múltipla – uma classe derivada é herdada de mais de uma classe base.
Q12. O python suporta herança múltipla?
    Resposta: Herança múltipla significa que uma classe pode ser derivada de mais de uma classe pai. Python suporta herança múltipla, ao contrário de Java.
